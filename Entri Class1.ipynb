```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/Users/sajan/employee data.csv')
df.info()

# load the data into a pandas DataFrame
data = pd.read_csv('employee data.csv')

# Task 1: Distribution of employees across teams
team_counts = data['Team'].value_counts()
total_employees = len(data)
team_percentages = (team_counts / total_employees) * 100

# Visualize distribution across teams
team_counts.plot (kind='bar',title='Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()

# Task 2: Segregate employees based on positions
position_counts = data['Position'].value_counts()

# Visualize distribution across Positions
position_counts.plot(kind='bar',title='Distribution of Employees Across Positions')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()

# Task 3: Predominent age group
age_groups = pd.cut(data['Age'],bins=[20,25,30,35,40],labels=['20-24','25-29','30-34','35-39'])
age_group_counts = age_groups.value_counts()

# Visualize age distribution
age_group_counts.plot(kind='bar', title='Age Distribution of Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()

# Task 4: Team and position with highest salary expenditure
team_salaries = data.groupby('Team')['Salary'].sum()
position_salaries = data.groupby('Position')['Salary'].sum()

# Visualize salary expenditure by team
team_salaries.plot(kind='bar', title='Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.show()

# Visualize salary expenditure by position
position_salaries.plot(kind='bar', title='Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.show()

# Task 5: correlation between age and salary
plt.scatter(data['Age'], data['Salary'],alpha=0.5)
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()

correlation = data['Age'].corr(data['Salary'])
print(f'Correlation between age and salary: {correlation:.2f}')
```
